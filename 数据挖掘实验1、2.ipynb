```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 设置中文字体和负号显示
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 加载数据集
iris = sns.load_dataset('iris')

# 基本统计描述
print("=== 数据集基本信息 ===")
print(iris.info())
print("\n=== 前5行数据 ===")
print(iris.head())

# 数值分析
numeric_cols = iris.select_dtypes(include=['float64'])
print("\n=== 数值特征分析 ===")
print("集中趋势指标:")
print(numeric_cols.describe().round(2))
print("\n离散趋势指标:")
print("标准差:\n", numeric_cols.std().round(2))
print("\n极差:\n", (numeric_cols.max() - numeric_cols.min()).round(2))

# 可视化部分
plt.figure(figsize=(10, 6))
sns.boxplot(data=numeric_cols)
plt.title('Iris数据集箱线图(数值特征)')
plt.tight_layout()
plt.savefig('iris_boxplot.png', dpi=300)
plt.show()

numeric_cols.hist(figsize=(12, 8), bins=15)
plt.suptitle('Iris数据集直方图(数值特征)')
plt.tight_layout()
plt.savefig('iris_histogram.png', dpi=300)
plt.show()

sns.pairplot(iris, hue='species', height=2.5)
plt.suptitle('Iris数据集散点图矩阵', y=1.02)
plt.tight_layout()
plt.savefig('iris_pairplot.png', dpi=300)
plt.show()
```

```python
from apyori import apriori

transactions = [
    ['面包', '牛奶'],
    ['面包', '尿布', '啤酒', '鸡蛋'],
    ['牛奶', '尿布', '啤酒', '可乐'],
    ['面包', '牛奶', '尿布', '啤酒'],
    ['面包', '牛奶', '尿布', '可乐']
]

def analyze_itemset(target, results, transactions):
    """分析特定项集的函数"""
    for result in results:
        items = [item for item in result.items]
        if set(target) == set(items):
            support = result.support
            support_count = int(support * len(transactions))
            return {
                'itemset': items,
                'support_count': support_count,
                'support': round(support, 2)
            }
    return None

# 第一题（1）优化
print("=== 第一题（1）===")
target = ['啤酒', '尿布', '牛奶']
results = list(apriori(transactions, min_support=0.4, min_confidence=0.5))
analysis = analyze_itemset(target, results, transactions)
if analysis:
    print(f"项集: {analysis['itemset']}, 支持数: {analysis['support_count']}, 支持度: {analysis['support']}")

# 第一题（2）优化
print("\n=== 第一题（2）===")
results2 = list(apriori(transactions, min_support=0.6, min_confidence=0.5))
for result in results2:
    items = [item for item in result.items]
    print(f"{len(items)}-项集: {items}, 支持度: {round(result.support, 2)}")

# 第一题（3）优化
print("\n=== 第一题（3）===")
target = ['啤酒', '尿布', '牛奶']
target1 = ['尿布', '牛奶']
for rule in results:
    items = [item for item in rule.items]
    if set(target) == set(items):
        print(f"项集: {items}, 支持度: {round(rule.support, 2)}")
        for ordered_stat in rule.ordered_statistics:
            antecedent = list(ordered_stat.items_base)
            consequent = list(ordered_stat.items_add)
            if set(target1) == set(antecedent):
                print(f"规则: {antecedent} → {consequent}, 置信度: {round(ordered_stat.confidence, 2)}")
```

```python
from apyori import apriori

def analyze_association_rules(transactions, min_support, min_confidence, title=""):
    """关联规则分析通用函数"""
    if title:
        print(f"\n=== {title} ===")
    
    results = list(apriori(transactions, min_support=min_support, min_confidence=min_confidence))
    
    print(f"强关联规则(最小支持度{min_support}, 最小置信度{min_confidence}):")
    for rule in results:
        items = [item for item in rule.items]
        
        for ordered_stat in rule.ordered_statistics:
            if ordered_stat.confidence >= min_confidence:
                antecedent = list(ordered_stat.items_base)
                consequent = list(ordered_stat.items_add)
                if antecedent:
                    print(f"规则: {antecedent} → {consequent}, 置信度: {ordered_stat.confidence:.2f}, 提升度: {ordered_stat.lift:.2f}")

# 第二题分析
transactions = [
    ['网球拍', '网球','运动鞋'],
    ['网球拍', '网球'],
    ['网球拍'],
    ['网球','运动鞋','羽毛球'],
    ['网球拍', '网球']
]

analyze_association_rules(
    transactions=transactions,
    min_support=0.5,
    min_confidence=0.6,
    title="第二题"
)
```

```python
from apyori import apriori

transactions = [
    ['1','2','5'],
    ['2','4'],
    ['2','3'],
    ['1','2','4'],
    ['1','3'],
    ['2','3'],
    ['1','3'],
    ['1','2','3','5'],
    ['1','2','3']
]

def print_rules(results, min_support, min_confidence):
    """打印关联规则的通用函数"""
    print("\n强关联规则(满足最小支持度{:.2f}和最小置信度{:.2f}):".format(min_support, min_confidence))
    for rule in results:
        items = [item for item in rule.items]
        for ordered_stat in rule.ordered_statistics:
            if ordered_stat.confidence >= min_confidence:
                antecedent = list(ordered_stat.items_base)
                consequent = list(ordered_stat.items_add)
                if antecedent:  # 确保前件不为空
                    print("规则: {} → {}, 支持度: {:.2f}, 置信度: {:.2f}, 提升度: {:.2f}".format(
                        antecedent, consequent, rule.support, 
                        ordered_stat.confidence, ordered_stat.lift
                    ))

# 第三题分析
print("=== 第三题 ===")
min_support = 2/9
min_confidence = 0
results = list(apriori(transactions, min_support=min_support, min_confidence=min_confidence))
print_rules(results, min_support, min_confidence)
```
